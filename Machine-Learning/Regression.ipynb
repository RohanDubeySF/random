```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
```

```python
data=fetch_california_housing()
data.data
```

```python
x=pd.DataFrame(data.data,columns=data.feature_names)
x.drop(['Latitude','Longitude'],axis=1,inplace=True)
x
```

```python
y=pd.DataFrame(data.target,columns=['target'])
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=69)
```

```python
x_train
```

## Standardize data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(LinearRegression(),x_train,y_train,scoring='r2',cv=20)
scores
```

```python
scores.mean()
```

```python
from sklearn.metrics import r2_score
y_reg=reg.predict(x_test)
r2_score(y_test,y_reg)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
```

```python
from sklearn.model_selection import GridSearchCV
param={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,50,70,80,90]}
ridgecv=GridSearchCV(ridge,param,scoring='r2',cv=5,return_train_score=True)
ridgecv.fit(x_train,y_train)
```

```python
ridgecv.best_score_
```

```python
y_ridge=ridgecv.predict(x_test)
r2_score(y_test,y_ridge)
```

#  Lasso Regression


```python
from sklearn.linear_model import Lasso
lasso=Lasso()
```

```python
param={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,50,70,80,90]}
lassocv=GridSearchCV(lasso,param,scoring='r2',cv=5,return_train_score=True)
lassocv.fit(x_train,y_train)
```

```python
lassocv.best_score_
```

```python
y_lasso=lassocv.predict(x_test)
r2_score(y_test,y_lasso)
```
