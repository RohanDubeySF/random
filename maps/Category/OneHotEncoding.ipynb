#1. Encoding using Pandas

```python
ddd=pd.get_dummies(data,columns=['fuel','owner'])
ddd
```

#2. K-1 Encoding (Removing 1 column)

```python
pd.get_dummies(data,columns=['fuel','owner'],drop_first=True).replace({True:1,False:0})
```

#Count or frequency Encoding

```python
 df=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',usecols=[' State-gov',' Bachelors',  ' Never-married',' Adm-clerical', ' Not-in-family', ' White', ' Male', ' United-States'])
df.head()
```

```python
df.columns=['Employment','Degree','Status','Designation','Family_job','Race','Sex','Country']
df.head()
```

```python
df.Country.value_counts()
```

```python
df.Country.nunique() #counts number of unique values in a column
```

```python
df.Country.replace(df.Country.value_counts(),inplace=True)
```

```python
df.Country=df.Country.map(df.Country.value_counts())
```

```python
df.Country.nunique() , df.Country.value_counts()
```
