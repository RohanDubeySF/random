```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day25-normalization/wine_data.csv',usecols=[0,1,2])
data.columns=['Class label','Alcohol','Malic acid']
data.head()
```

```python
sns.kdeplot(data['Alcohol'])
```

```python
sns.kdeplot(data['Malic acid'])
```

```python
sns.scatterplot(data=data,x=data['Alcohol'],y=data['Malic acid'],hue=data['Class label'])
```

```python
color_dict={1:'red',3:'green',2:'blue'}
sns.scatterplot(data=data,x=data['Alcohol'],y=data['Malic acid'],hue=data['Class label'],palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(data[['Alcohol','Malic acid']],data['Class label'],test_size=0.3,random_state=0)
x_train.shape,x_test.shape
```

```python
 from sklearn.preprocessing import MinMaxScaler
 scaler=MinMaxScaler()
 scaler.fit(x_train)
```

```python
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=['Alcohol','Malic acid'])
x_test_scaled=pd.DataFrame(x_test_scaled,columns=['Alcohol','Malic acid'])
x_train_scaled
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(x_train_scaled.describe(),1)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2, figsize=(10,4))
color_dict={1:'red',3:'green',2:'blue'}
ax1.set(title='Before Scaling')
sns.scatterplot(data=x_train,x=x_train['Alcohol'],y=x_train['Malic acid'],hue=y_train,ax=ax1,palette=color_dict)

ax2.set(title='After Scaling')
sns.scatterplot(data=x_train_scaled,x=x_train_scaled['Alcohol'],y=x_train_scaled['Malic acid'],hue=y_train,ax=ax2,palette=color_dict)
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2, figsize=(10,4))

ax1.set(title="Before Scaling",xlabel=' ')
sns.kdeplot(x_train['Alcohol'],ax=ax1,palette='r')
sns.kdeplot(x_train['Malic acid'],ax=ax1,palette='g')

ax2.set(title="After Scaling")
sns.kdeplot(x_train_scaled['Alcohol'],ax=ax2)
sns.kdeplot(x_train_scaled['Malic acid'],ax=ax2)
plt.show()
```
